# 1. One-time investment with equal weights of stocks

**Sharp Ratio**

```python
def compute_sharpe_ratio(data, risk_free_rate = 0):
    mean_return = data["total"].mean()
    std = data["total"].std()
    sharpe_ratio = (mean_return-risk_free_rate) / std
    return sharpe_ratio
```

```python
dailySR = compute_sharpe_ratio(data=retail_returns)
MonthlySR = dailySR * (30.4**0.5)
yearlySR = dailySR * (365**0.5)
```

```python
print(f'Daily Sharp ratio for portfolio: {round(dailySR,2)}')
print(f'Monthly Sharp ratio for portfolio: {round(MonthlySR,2)}')
print(f'Yearly Sharp ratio for portfolio: {round(yearlySR,2)}')
```

```python
def compute_prob_sharpe_ratio(data, benchmark=0):
    
    sr = compute_sharpe_ratio(data, 0)
    
    skew = scipy.stats.skew(data["total"])
    # Use fisher kurtosis
    
    kurtosis = scipy.stats.kurtosis(data["total"], fisher=True)  
    
    n = len(data)
    
    std = ( (1 / (n-1)) * (1 + 0.5 * sr**2 - skew * sr + (kurtosis / 4) * sr**2))**0.5
    
    ratio = (sr - benchmark) / std
    prob_sharpe_ratio = scipy.stats.norm.cdf(ratio)
    return prob_sharpe_ratio
```

```python
DaylyProbaSH = compute_prob_sharpe_ratio(data=retail_returns)
```

```python
YearlyProbaSH = DaylyProbaSH*(365**0.5)
```

```python
print(f'Daily Probabiliscic Sharp ratio for portfolio: {round(DaylyProbaSH,2)}')
print(f'Yearly Probabiliscic Sharp ratio for portfolio: {round(YearlyProbaSH,2)}')
```
